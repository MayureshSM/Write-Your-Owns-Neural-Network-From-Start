```python
#Import numpy and pandas
import numpy as np
import pandas as pd

# Define the Traing data, Test data and NN weghts as global
global XTrain,YTrain,XTest,YTest,Twt
```

```python
#Class is defined to store the nodes values/weights of each layer
class Layer():
    def __init__(self,):
        self.nodes=''
```

```python
#Define an activation function of the NN. This could be Relu or sigmoid function

def Activation(d,f):
    if f=='Sigmoid':
        d=1/(1+np.exp(-d))
    elif f=='Relu':
        ind=np.where(d<0)
        d[ind]=0
    return(d)
```

```python
#Define forward propogation of the NN
def Forward_propogation(D,wt1):
    X[0].nodes=D
    for i in range(len(wt1)):
        D=np.matmul(D,wt1[i].nodes)
        D=Activation(D,'Relu')
        X[i+1].nodes=D
    return(D[0])
```

```python
#Define the error/cost functon of the the NN
def Cost_function(wt1):
    yp=[]
    for i in range(len(XTrain)):
        yp.append(Forward_propogation(np.array([XTrain[i]]),wt1))
    cost=np.sum((yp-YTrain)**2)
    return(cost)
```

```python
#Train the model using the gradend descent method to refine he weights of each layer
def Train():
    global Twt
    cost3=0
    
    while True:
        for i in range(len(LN)-1):
            dwt[i].nodes=0*dwt[i].nodes
        #Find the cost function 
        cost1=Cost_function(Twt)
        if abs(cost1-cost3)<100000:
            print(cost3)
            break
        else:
            for i in range(len(Twt)):
                for j in range(len(Twt[i].nodes)):
                    for k in range(len((Twt[i].nodes).T)):
                        a=Twt[i].nodes[j,k]+0
                        Twt[i].nodes[j,k]=Twt[i].nodes[j,k]+0.1
                        cost2=Cost_function(Twt)
                        dwt[i].nodes[j,k]=(cost2-cost1)/0.1
                        Twt[i].nodes[j,k]=a
        dwt2=0
        for i in range(len(dwt)):
            dwt2=np.sum((dwt[i].nodes)**2)+dwt2
        dwt2=np.sqrt(dwt2)
        for i in range(len(dwt)):
            dwt[i].nodes=dwt[i].nodes/dwt2
            Twt[i].nodes=-0.1*dwt[i].nodes+Twt[i].nodes
        cost3=cost1+0
        print(cost3)
```
